```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def pltlines(folder_dir, axname=None):
    df1 = pd.read_csv(f"../test_results/{folder_dir}/0.csv")
    df2 = pd.read_csv(f"../test_results/{folder_dir}/96.csv")


    df1["Prediction"] = None
    tmp_lst = list(df1["Prediction"])
    tmp_lst[-1] = list(df1["Ground Truth"])[-1]
    df1["Prediction"] = tmp_lst
    df3 = pd.concat([df1, df2], ignore_index=True).reset_index()
    df3 = df3.rename(columns={"index": "Time Steps"})
    df3 = pd.melt(df3, id_vars=['Time Steps'], var_name='Data Type')
    ax = sns.lineplot(data=df3, x="Time Steps", y="value", hue="Data Type", ax=axname)
    ax.get_legend().set_title("")
    return ax
```

```python
fig, axes = plt.subplots(3, 1, sharex=True, figsize=(8, 6))

# Adding each seaborn plot to the corresponding axis
plot2 = pltlines("long_term_forecast_Mamba-130m_ETTh1_pl720_sl672_ll576_tl96_lr0.0001_bt64_wd0_cosFalse_test__0", axes[0])
axes[0].set_title("ETTh1")
plot1 = pltlines("long_term_forecast_Mamba-130m_custom_pl720_sl672_ll576_tl96_lr0.0001_bt64_wd0_cosFalse_test_traffic2_0", axes[1])
axes[1].set_title("Traffic")
plot3 = pltlines("long_term_forecast_Mamba-130m_custom_pl720_sl672_ll576_tl96_lr0.0001_bt64_wd0_cosFalse_test_electricity1_0", axes[2])
axes[2].set_title("Electricity")
#plot4 = pltlines("long_term_forecast_gpt2_custom_pl720_sl672_ll576_tl96_lr0.0001_bt64_wd0_cosFalse_test_1", axes[1][1])
```
